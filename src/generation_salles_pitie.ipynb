```python
# Salles d'isolement
import numpy as np  
import pandas as pd
from datetime import datetime, timedelta
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# dates
date_debut = datetime(2024, 1, 1)
date_fin = datetime(2024, 12, 31)
date_range = pd.date_range(date_debut, date_fin, freq='D')


salles_isolement = []
poules_isolation = {
    'Réanimation': {'salles': 12, 'lits_salle': 2, 'equip_ventil': 24},
    'Maladies_infectieuses': {'salles': 8, 'lits_salle': 3, 'equip_ventil': 18}, 
    'Urgences': {'salles': 6, 'lits_salle': 2, 'equip_ventil': 12},
    'Médecine_interne': {'salles': 4, 'lits_salle': 2, 'equip_ventil': 8}
}

for date in date_range:
    for pole, config in poules_isolation.items():
        occupation_base = np.random.normal(0.65, 0.15)
        if date.month in [12,1,2]:  # Hiver
            occupation_base *= 1.35
        elif date.month in [7,8]:  # Été
            occupation_base *= 1.20
            
        nb_lits_occupes = int(config['lits_salle'] * config['salles'] * occupation_base)
        
        salles_isolement.append({
            'date': date,
            'pole': pole,
            'nb_salles': config['salles'],
            'lits_total': config['lits_salle'] * config['salles'],
            'lits_occupes': max(0, nb_lits_occupes),
            'taux_occupation': min(1.0, occupation_base),
            'ventilateurs': config['equip_ventil'],
            'ventil_dispo': int(config['equip_ventil'] * (1-occupation_base)),
            'epidemic_risk': 1 if occupation_base > 0.95 else 0
        })

df_isolement = pd.DataFrame(salles_isolement)
df_isolement.to_csv('data/salles_isolement_pitie.csv', index=False)
print(f"salles_isolement_pitie.csv généré ({len(df_isolement):,} lignes)")
```
